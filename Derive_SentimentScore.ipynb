```python
import collections

data_dict = {
'APPLE':'AAPL',
'MICROSOFT':'MSFT',
'NVDIA':'NVDA',
'IBM':'IBM',
'ADOBE':'ADBE',
'ALPHABET':'GOOGL',
'GOOGLE':'GOOG',
'FACEBOOK':'FB',
'MASTERCARD' : 'MA',
'VISA':'V',
'CISCO':'CSCO',
'ORACLE':'ORCL',
'PAYPAL':'PYPL',
'ADP':'ADP',
'ACTIVISION':'ATVI',
'HP':'HP',
'INTEL':'INTC'
}

#[{Date:'',AAPL:'22.5'},{}]

arrFinalScores = []

#dictFinal_Sentiments = {}

#dictFinal_Scores = {}

for filePathFull in Final_file_Paths:
    #dictFinal_Sentiments = {}
    dictFinal_Sentiments = collections.OrderedDict()
    newFilePathFull = filePathFull.split('\n')[0]
    arrKey = newFilePathFull.split('/')[6]
    strDateKey = newFilePathFull.split('/')[7].split('_')[2].split('.')[0]

    print('File Path is :'+newFilePathFull)
    print('The Ticker Symbol is : '+ arrKey.upper())
    fileOpen = open(newFilePathFull,'r')
    file_contents = fileOpen.read()
    arrNew3 = file_contents.split('\n')
    arrNew3 = list(filter(None, arrNew3)) 
    chunks = [arrNew3[x:x+25] for x in range(0, len(arrNew3), 25)]
    response_list = []
    for listItem in chunks:
        if(len(listItem)) >= 1:
            response = comprehend.batch_detect_sentiment(
            TextList=listItem,
            LanguageCode='en'
            )
            response_list.append(response)
    total_positive_score = 0.0
    total_negative_score = 0.0
    for item_list in response_list:
        for dict_responses in item_list['ResultList']:
            if dict_responses['SentimentScore']:
                total_positive_score += dict_responses['SentimentScore']['Positive']
                total_negative_score += dict_responses['SentimentScore']['Negative']         
    final_score = total_positive_score - total_negative_score
    strTickerSymbol = data_dict[arrKey.upper()]
    dictFinal_Sentiments[strTickerSymbol] = final_score
    dictFinal_Sentiments['Date'] = strDateKey
    arrFinalScores.append(dictFinal_Sentiments)
    
```
